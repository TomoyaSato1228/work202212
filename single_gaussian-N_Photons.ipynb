```python
#FWHMを変化させて,それぞれのmeanをlistに入れて標準偏差を導出(FWHM補正)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from matplotlib import pyplot as plt
N=15#FWHM_range
P=100#Photons
S=1000#number_of_step
FWHM_list=[]
for FWHM in range(1,N+1):
    sigma_list=[]
    FWHM_list.append(FWHM)
    print("FWHM = ", FWHM)
    mean_list=[]
    photon_list=[]
    for s in range(10,P+1):
        photon_list.append(s)
        for n in range(1,S+1):
            a=np.random.normal(
                loc   = 0,
                scale = FWHM*0.42466452,
                size  = s,
                )
            mean=np.mean(a)
            mean_list.append(mean)
            #print("n = ", n)
            #print("mean = ", mean)
    #print(mean_list)
        st_d=np.std(mean_list)
    #print("Standard deviation = ", st_d)
    #print("2σ = " , st_d*2)
        sigma_list.append(st_d*2)
    #print(sigma_list)
    #print(photon_list)
    #plt.xscale("log")
    plt.plot(photon_list,sigma_list)
```

```python
print(FWHM_list)
plt.scatter(FWHM_list,sigma_list)
plt.xlabel("FWHM [arcmin]")
plt.ylabel("2σ [arcmin]")
plt.grid()

linear_model=np.polyfit(FWHM_list,sigma_list,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,N+1)
plt.plot(x_s,linear_model_fn(x_s),color="red")
print(linear_model_fn)
```
