```python
#それぞれのmeanを導出
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
for n in range(1,6):
    a=np.random.normal(
    loc   = 0,
    scale = 1,
    size  = 10,
        )
    mean=np.mean(a)
    print("n = ", n)
    #print("a = ", a)
    print("mean = ", mean)
```

```python
#FWHMを変化させてmeanを導出
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
for FWHM in range(1,6):
    a=np.random.normal(
    loc   = 0,
    scale = FWHM,
    size  = 10,
        )
    mean=np.mean(a)
    print("FWHM = ", FWHM)
    print("a = ", a)
    print("mean = ", mean)
```

```python
#FWHMを変化させて5回ずつmeanを導出
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
for FWHM in range(1,6):
    print("FWHM = ", FWHM)
    for n in range(1,6):
        a=np.random.normal(
            loc   = 0,
            scale = FWHM,
            size  = 10,
            )
        mean=np.mean(a)
        print("n = ", n)
        print("a = ", a)
        print("mean = ", mean)
```
