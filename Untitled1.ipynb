```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
plt.rcParams['font.size'] = 14
import numpy as np


b = np.random.normal(
    loc   = 0,      # 平均
    scale = 1,    # 標準偏差
    size  = 10000,# 出力配列のサイズ(タプルも可)
)

param = norm.fit(b)
print (param)

hist_1, bins = np.histogram(b, 100, range=(-5,5))
bins = bins[:-1]

def func(x, a, mu, sigma):
    return a*np.exp(-(x-mu)**2/(2*sigma**2))

param_ini = [1000,0,2]
popt, pcov = curve_fit(func, bins, hist_1, p0=param_ini)
print(popt)
fitting = func(bins, popt[0],popt[1],popt[2])


fig, ax = plt.subplots()
ax.bar(bins,hist_1,width=0.1, alpha=0.5,color='m',align='edge')
ax.plot(bins,fitting,'k')
ax.set_xlabel('x')
ax.set_ylabel('y')
#plt.savefig('curvefit2.png', dpi=120) 　　#グラフの画像を保存したいときに#を外す
plt.show()
```

```python

import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
plt.rcParams['font.size'] = 14
import numpy as np

#ガウス分布にしたがう乱数の設定
sigma=1
N=100

counts_list=[]

sigma_res_list=[]

for i in range (1000):

    b = np.random.normal(
        loc   = 0,      # 平均
        scale = sigma,    # 標準偏差
        size  = N,# 出力配列のサイズ(タプルも可)
    )

    param = norm.fit(b)
    print (param)

    #sample_1 = norm.rvs(loc=100,scale=5,size=1000)
    hist_1, bins = np.histogram(b, 100, range=(-50,50))   #生成したガウシアンのヒストグラム
    bins = bins[:-1]


    def func(x, a, mu, sigma):
        return a*np.exp(-(x-mu)**2/(2*sigma**2))


    param_ini = [1000,0,2]
    popt, pcov = curve_fit(func, bins, hist_1, p0=param_ini)
    #print(popt)
    fitting = func(bins, popt[0],popt[1],popt[2])
    sigma_res=popt[2]-sigma
    sigma_res_list.append(sigma_res)
    total_counts= popt[0]*np.sqrt(np.pi)*(np.exp(1/(2*popt[2]**2)))

    #print("total counts =", total_counts)
    
    counts_list.append(total_counts)


    #fig, ax = plt.subplots()
    #ax.bar(bins,hist_1,width=1, alpha=0.7,color='m',align='edge')
    #ax.plot(bins,fitting,'k')
    #ax.set_xlabel('x')
    #ax.set_ylabel('y')
    #plt.savefig('curvefit2.png', dpi=120)
    #plt.show()

#plt.hist(b,bins=100)
#plt.xlim(-10,10)

plt.hist(counts_list, bins=25, range =(50, 150))
plt.title("N")
plt.show()

ret=plt.hist(sigma_res_list, bins=50, range =(-1, 1))
plt.title("Δσ")
plt.show()
```
