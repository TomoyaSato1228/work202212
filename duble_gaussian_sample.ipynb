```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
N=10000
ni=[1.0,0.7,0.5,0.3,0.]
FWHM=8

for d in range(1):
    mean_list=[]
    for i in range(1):
        A = []
        for n in range(len(ni)):
            n1=int(N*ni[n])
            n2=N-n1
            a1=np.random.normal(
                loc   = 0,
                scale = FWHM/2.35,
                size  = n1,
                )
            #print("a1= ", a1)
            a2=np.random.normal(
                loc   = d+5,
                scale = FWHM/2.35,
                size  = n2,
                )
            #print("a2 ", a2)
            A=np.append(a1,a2)
            #print(A)
            mean=np.mean(A)
            mean_list.append(mean)
            print("n1 = ", n1)
            print("n2 = ", n2)
            print("mean = ", mean)
            #plt.hist(a1)
            plt.hist(A,bins=50,histtype="step",label="gaus1:gaus2 = {:.1f} : {:.1f}".format(ni[n], 1-ni[n]))
            plt.xlabel("[arcmin]")
            plt.ylabel("number of trials")
            plt.legend(bbox_to_anchor=(1.05,0.9))


            
    print(mean_list)
#st_d=np.std(l)
#print("Standard deviation = ", st_d)
```
